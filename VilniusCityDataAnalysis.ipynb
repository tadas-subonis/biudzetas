```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0/1000))
        return '{v:d}k €'.format(v=val)
    return my_autopct

data = pd_saskaitos_s.query('PAYMENT_DATE > "2018"').groupby(pd.Grouper(key='TIEKEJAS')) \
    .agg({'Išmokėta':'sum'}) \
    .nlargest(10, 'Išmokėta')
    


data['Išmokėta'].plot.pie(title='Vilniaus miesto savivaldybės išlaidų TOP10, 2018 m.', autopct=make_autopct(data['Išmokėta']))
```
